```python
# Simple Thresholding
```

```python
import cv2
```

```python
#read image

image=cv2.imread("car.jpg",0)   # 0-Grayscale image 1-COlor image
```

```python
# Apply Binary Threshold -Simple Thresholding have One fixed threshold used for Clear images

rest, threshold=cv2.threshold(image,165,255,cv2.THRESH_BINARY)
```

```python
#Show image

cv2.imshow("Original image:",image)
cv2.imshow("Simple Threshold",threshold)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Adavtive threshold - Adaptive Thresholding have Different thresholds for regions and used for Uneven lighting

adap=cv2.adaptiveThreshold(
    image,
    255,
    cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
    cv2.THRESH_BINARY,
    11,
    2)
```

```python
#Show image

cv2.imshow("Original image:",image)
cv2.imshow("Adaptive Threshold",adap)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#otsu threshold - Otsu Thresholding have Automatic threshold selection used for Complex images
```

```python
ret, otsu=cv2.threshold(
    image,
    0,
    255,
    cv2.THRESH_BINARY + cv2.THRESH_OTSU)
```

```python
#Show image

cv2.imshow("Original image:",image)
cv2.imshow("Otsu Threshold",otsu)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
